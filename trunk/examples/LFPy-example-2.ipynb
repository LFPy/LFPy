# Example 2: Extracellular response of a single synapse #First step is to **import LFPy** and other packages:

```python
import LFPyfrom pylab import *
```

Create some dictionarys with parameters for cell, synapse and extracellular electrode:

```python
cellParameters = {    'morphology' : 'morphologies/L5_Mainen96_wAxon_LFPy.hoc',}synapseParameters = {    'e' : 0.,    'syntype' : 'Exp2Syn',    'tau1' : 0.5,    'tau2' : 2.0,    'weight' : 0.005,}electrodeParameters = {    'x' : zeros(101),    'y' : zeros(101),    'z' : arange(101) * 10,    'sigma' : 0.3,}
```

Then, create the **cell**, **synapse** and **electrode** objects using **LFPy.Cell**, **LFPy.Synapse**, **LFPy.RecExtElectrode**.

```python
cell = LFPy.Cell(**cellParameters)cell.set_pos(xpos=10, ypos=0, zpos=0)synapse = LFPy.Synapse(cell,                       idx = cell.get_closest_idx(z=800),                       **synapseParameters)synapse.set_spike_times(array([10, 30, 50]))                        electrode = LFPy.RecExtElectrode(**electrodeParameters)
```

Run the simulation with **electrode** as input to **cell.simulate()**

```python
cell.simulate(electrode = electrode)
```

Then plot the **somatic potential** and the **LFP**from electrode object:

```python
figure(figsize=(12, 6))subplot(121)pcolormesh(cell.tvec, electrode.z, electrode.LFP), colorbar(), title('LFP (mV)')subplot(122)plot(cell.tvec, cell.somav), title('somatic voltage (mV)')show()
```
