# Example 2: Extracellular response of a single synapse #First step is to **import LFPy** and other packages:

```python
import LFPyfrom pylab import *
```

Then, create the cell object using the **LFPy.Cell**-class, specifying the morphology file.The passive mechanismsare switched on by default.

```python
cell = LFPy.Cell(morphology = 'morphologies/L5_Mainen96_wAxon_LFPy.hoc')
```

One can now use **LFPy.Synapse**-class to insert a single synapse onto the soma compartment, and set the spike time(s) using **LFPy.Synapse.set_spike_times()**:

```python
synapse = LFPy.Synapse(cell,                        idx = cell.get_closest_idx(z=800),                       syntype = 'Exp2Syn',                        weight = 0.005,                        e = 0,                        tau1 = 0.5,                       tau2 = 2,                       record_current = False)synapse.set_spike_times(array([20., 40]))
```

We now create the extracellular electrode, using class **LFPy.RecExtElectrode**

```python
electrodeParams = {                   'x' : zeros(101)+10,                   'y' : zeros(101),                   'z' : arange(101) * 10,                   'sigma' : 0.3,                   }electrode = LFPy.RecExtElectrode(**electrodeParams)
```

We now have what we need to calculate the postsynaptic response, without calculating LFPs,using a built in method **LFPy.Cell.simulate()** method to run the simulation.The **electrode**-object should be given as argument to **cell.simulate()**.

```python
cell.simulate(electrode = electrode)
```

Then plot the somatic potential and the LFP from electrode object:

```python
figure(figsize=(12, 6))subplot(121)pcolormesh(cell.tvec, electrode.z, electrode.LFP), colorbar(), title('LFP (mV)')subplot(122)plot(cell.tvec, cell.somav), title('somatic voltage (mV)')show()
```
